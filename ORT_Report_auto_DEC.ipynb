For Finding ORT manual date sent by the system, The system follows a specific template and the index of the maual_ORT_date in the system's reply can be found from the string and once it is found, it is compared with the most recent ORT_manual date and if it matches, it is considered as a manual_ORT_OK case.

```python
ext_inf=df['External Info']
dte=df['ORT manual']
```

```python
droplst=[]
count=0
for i in range(0,len(df)):
  ext_check = word_tokenize(str(ext_inf[i]))
  dte_check = word_tokenize(str(dte[i]))
  for j in ext_check:
    if j==dte_check[0]:
      count=count+1
      droplst.append(i)
      break
```

Following is the code to check the ORT manual date if it is sent by the case processor and not the system since it can be anywhere in the reply.
( depends on the template used by the case processor ).

```python
Userdf = df
for i in droplst:
  Userdf.drop(i,axis=0,inplace=True)
```

```python
Userdf
```

```python
ext_inf=Userdf['External Info']
dte=Userdf['ORT manual']
```

```python
def is_valid_date(date_str, format='%Y-%m-%d'):
    try:
        datetime.strptime(date_str, format)
        return True
    except ValueError:
        return False

count2=0
dtelst=[]
for i in range(0,len(Userdf)):
  test=Userdf.iloc[i,1]
  for j in test:
    x=test.index(j)
    if j.isdigit():
      dtestr=test[x:(x+10)]

    if is_valid_date(dtestr):
      dtelst.append(dtestr)
    else:
      pass
  word_lst = word_tokenize(str(Userdf.iloc[i,2]))
  if word_lst[0] in dtelst:
    count2=count2+1
    continue
```
