```python
def is_date(string):
    try:
        parse(string)
        return True
    except ValueError:
        return False


ext_inf=df['External Info']
dte=df['ORT manual']
count=0

for i in range(0,(len(df)-1)):
  ext_check = word_tokenize(str(ext_inf[i]))
  dte_check = word_tokenize(str(dte[i]))
  for j in ext_check:
    if is_date(j):
      if check_dte(str(j),str(dte_check[0])):
        count=count+1
      else:
        pass

print(count)
```
