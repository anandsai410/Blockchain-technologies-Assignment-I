```python
import nltk
import datetime
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
nltk.download('punkt')
from nltk import punkt
import pandas as pd
import numpy as np
import re
from dateutil import parser
from datetime import datetime
```

```python
df = pd.read_excel("cust_Test.xlsx")
df
```

```python
def convert_dte(x):
  if (x[0:4]+x[5:7]+x[8:10]).isdigit() and x[4]=="-":
    y=datetime.strptime(x, "%Y-%m-%d")
    y=y.strftime("%Y.%m.%d")
    return y
  if (x[0:2]+x[3:5]+x[6:10]).isdigit() and x[2]=="-":
    y=datetime.strptime(x, "%d-%m-%Y")
    y=y.strftime("%Y.%m.%d")
    return y
  if (x[0:4]+x[5:7]+x[8:10]).isdigit() and x[4]==".":
    y=datetime.strptime(x, "%Y.%m.%d")
    y=y.strftime("%Y.%m.%d")
    return y
  if (x[0:2]+x[3:5]+x[6:10]).isdigit() and x[2]==".":
    y=datetime.strptime(x, "%d.%m.%Y")
    y=y.strftime("%Y.%m.%d")
    return y
  if (x[0:4]+x[5:7]+x[8:10]).isdigit() and x[4]=="/":
    y=datetime.strptime(x, "%Y.%m.%d")
    y=y.strftime("%Y.%m.%d")
    return y
  if (x[0:2]+x[3:5]+x[6:10]).isdigit() and x[2]=="/":
    y=datetime.strptime(x, "%d.%m.%Y")
    y=y.strftime("%Y.%m.%d")
    return y

def repeat_check(a,b):
  a=parser.parse(a)
  b=parser.parse(b)
  if (a<b):
    return True

def check_dte(a,b):
  a=parser.parse(a)
  b=parser.parse(b)
  if a==b:
    return True

def is_date(string):
    try:
      parser.parse(string)
      return True
    except ValueError:
      return False
```

```python
ext_inf=df['External Info']
dte=df['ORT manual']
countok=0
rcount=0
countnok=0

curr_y = datetime.now().year

for i in range(0,(len(df)-1)):
  ext_check = word_tokenize(str(ext_inf[i]))
  dte_check = word_tokenize(str(dte[i]))
  for j in ext_check:
    if is_date(j):
      if (j[0:2]+j[3:5]+j[6:10]).isdigit() and len(j)>9:
        j=convert_dte(j)
      if check_dte(str(j),str(dte_check[0])):
        countok=countok+1
        continue
      if repeat_check(str(j),str(dte_check[0])) and j!="2023": #j!=curr_y:
        rcount=rcount+1

countnok=len(df)-(countok+rcount)
print(countok)
print(rcount)
print(countnok)
```
